```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn import model_selection
```

```python
df_car = pd.read_csv('automobile_dataset_preprocessing.csv')
pd.set_option('display.max_columns', None)
df_car.head()
```

```python
df_car = df_car.drop('Unnamed: 0', axis=1)
df_car.head()
```

```python
df_car.describe()
```

```python
# menghapus(memisahkan) kolom price / mengambil kolom-kolom attribute
df_carFeatures = df_car.drop(['price'], axis=1)
df_carFeatures.head()
```

```python
# menghapus(memisahkan) kolom Class dan Gender / mengambil kolom yang berisi numerik
df_numeric_features = df_carFeatures.drop(['make', 'fuel-type', 'aspiration', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'fuel-system', 'price-binned'], axis=1)
df_numeric_features.head()
```

```python
df_numeric_features.describe()
```

```python
df_carEncoding = pd.get_dummies(df_carFeatures)
df_carEncoding.head(10)
```

```python
df_carJoin = pd.concat([df_carEncoding, df_car['price']], axis=1)
df_carJoin.head()
```

```python
source = df_carJoin.iloc[:,:-1] 
target = df_carJoin['price'] 

sv_train, sv_test, tv_train, tv_test = train_test_split(source, target, random_state=3, test_size=0.2)
print("\nData Training:\n")
print(sv_train.head())
print(sv_train.shape)

print("\nData Testing:\n")
print(sv_test.head())
print(sv_test.shape)
```

```python
regressor_model = DecisionTreeRegressor(random_state = 0)
regressor_model.fit(sv_train, tv_train)
```

```python
pred = regressor_model.predict(sv_test)
pred
```

```python
results = model_selection.cross_val_score(regressor_model, source, target, cv=4, scoring='r2')
print(results.mean())
```

```python
score=regressor_model.score(sv_test, tv_test)
print(score)

mse =mean_squared_error(tv_test, pred)
print("Mean Squared Error:",mse)

rmse = math.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```
